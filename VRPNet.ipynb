# Training 

```python
def train_model(model, trainloader, testloader, class_weights, epochs, beam_size=100, num_vehicles=5, lr=0.001):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    criterion = nn.CrossEntropyLoss(weight=class_weights)
    
    n_iters = model.n_iters
    
    batch_losses = []
    losses = []
    avg_opt_gaps = []
    
    for ep in range(epochs):
        model.train()
        running_loss = 0
        for node_inputs, edge_inputs, targets, dist_mat, route_lengths, src_ids, dst_ids in iter(trainloader): 

            outputs = model(node_inputs, edge_inputs, src_ids, dst_ids)            
            outputs = outputs.flatten(start_dim=1, end_dim=2)
            outputs = outputs.permute(1, 2, 0)

            targets = targets.expand(n_iters, -1).permute(1, 0)

            loss = criterion(outputs, targets)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            batch_losses.append(loss.item())
            
            running_loss += loss.item() * batch_size

        losses.append(running_loss / (len(trainloader) * batch_size))

        avg_opt_gap, avg_pred_route_len = average_beamsearch_optimality_gap(model, 
                                                                            testloader, 
                                                                            beam_size, 
                                                                            num_vehicles, 
                                                                            device, 
                                                                            shortest=False)

        avg_opt_gaps.append(avg_opt_gap.cpu())

        print(f"Epoch {ep+1} completed, with train loss {losses[ep]}, and" )
        print(f"                        average optimality gap {avg_opt_gap}")
        print()
        
    return losses, avg_opt_gaps, batch_losses
```
